````python
# PART I
# This will generate testvectors with expexted values
#ONE BIT


import numpy as np

A = []
B = []
Cin = []
Cout = []
Sum = []
Total = []

a = np.genfromtxt("C:\\Users\\noah_\\Downloads\\PYTHON\\March_Mini_Project\\March_Mini_Project\\RCA_1_Bit_HW.txt",dtype = int, unpack = True)
b = np.delete(a,0,1)
c = np.transpose(b)
d = c.flatten()


length = int(len(d)/np.size(c,0))


A.append(d[0::length])
B.append(d[1::length])
Cin.append(d[2::length])


def FullAdder(A,B,Cin):
    Cout = (A^B and Cin) or (A and B)
    Sum = (A^B)^Cin
    Total = (2*Cout)+Sum

    return Cout,Sum, Total


for i in range(np.size(A)):

    x,y,z = FullAdder(A[0][i],B[0][i],Cin[0][i])
    Cout.append(x)
    Sum.append(y)
    Total.append(z)

def writeOut(A,B,Cin,Cout,Sum,Total):
    
    outF = open("RCA_1_Bit.txt", "w")
    
    #giving values to text required for .txt doc
    l ='A';
    m ='B';
    n ='Cin';
    o ='Cout';
    p = 'Sum';
    q ='Total';
 
    #formatting .txt doc
    outF.write(f'{l:^4} {m:^4} {n:^4} {o:^4} {p:^4} {q:^4} \n')
    temp1 = []
    temp2 = []    
    temp3 = []    
    temp4 = []  
    temp5 = []
    temp6 = []
    for i in range(np.size(A)):
 
        temp1.append(str(A[0][i]))
        temp2.append(str(B[0][i]))
        temp3.append(str(Cin[0][i]))
        temp4.append(str(Cout[i]))
        temp5.append(str(Sum[i]))
        temp6.append(str(Total[i]))
     
    for i in range(np.size(A)):
        outF.write(f'{temp1[i]:^4} {temp2[i]:^4} {temp3[i]:^4} {temp4[i]:^4} {temp5[i]:^4} {temp6[i]:^4} \n')

    
    outF.close()
    
    return

writeOut(A,B,Cin,Cout,Sum,Total)
````

````python
# PART I
# This will generate testvectors with expexted values
#4 BIT
import numpy as np
import math as m

A = []
B = []
Cin = []
Total = []

Cout = []


b_10 = []
b_10a = []
b_10b = []
b_10c = []
b_10d = []
b_10_convert = []


a = np.genfromtxt("C:\\Users\\noah_\\Downloads\\PYTHON\\March_Mini_Project\\March_Mini_Project\\RCA_4_Bit_HW.txt",dtype = int, unpack = True)
b = np.delete(a,0,1)
c = np.transpose(b)
d = c.flatten()

w1 = []
w2 = []
x1 = []
x2 = []
y1 = []
y2 = []
z1 = []
z2 = []

ww1 = []
ww2 = []
xx1 = []
xx2 = []
yy1 = []
yy2 = []
zz1 = []
zz2 = []


length = int(len(d)/np.size(c,0))


A.append(d[0::length])
B.append(d[1::length])
Cin.append(d[2::length])

   
for i in range(np.size(A)):
    w1.append(A[0][i]%2)
    w2.append(m.floor(A[0][i]/2))
    
    x1.append(w2[i]%2)
    x2.append(m.floor(w2[i]/2))
    
    y1.append(x2[i]%2)
    y2.append(m.floor(x2[i]/2))
    
    z1.append(y2[i]%2)
    z2.append(m.ceil(y2[i]/2)) # this all is A

    ww1.append(B[0][i]%2)
    ww2.append(m.floor(B[0][i]/2))
    
    xx1.append(ww2[i]%2)
    xx2.append(m.floor(ww2[i]/2))
    
    yy1.append(xx2[i]%2)
    yy2.append(m.floor(xx2[i]/2))
    
    zz1.append(yy2[i]%2)
    zz2.append(m.ceil(yy2[i]/2)) # this all is B
    
    
for i in range(np.size(A)):   
        
    Cout1 = (((w1[i]^ww1[i]) and Cin[0][i]) or (w1[i] and ww1[i]))
    Sum1 = w1[i]^ww1[i]^Cin[0][i]

    Cout2 = (((x1[i]^xx1[i]) and Cout1) or (x1[i] and xx1[i]))
    Sum2 = x1[i]^xx1[i]^Cout1

    Cout3 = (((y1[i]^yy1[i]) and Cout2) or (y1[i] and yy1[i]))
    Sum3 = y1[i]^yy1[i]^Cout2

    Cout4 = (((z1[i]^zz1[i]) and Cout3) or (z1[i] and zz1[i]))
    Sum4 = z1[i]^zz1[i]^Cout3
        
    b_10d.append(Sum4)
    b_10c.append(Sum3)
    b_10b.append(Sum2)
    b_10a.append(Sum1)
        
    Cout.append(Cout4)
        
for i in range(np.size(A)): # SUM: converting binary to Decimal
    b_10_convert.append((b_10d[i]*(2**3))+(b_10c[i]*(2**2))+(b_10b[i]*(2**1))+(b_10a[i]*(2**0)))


for i in range(np.size(A)):
    Total.append(A[0][i]+B[0][i]+Cin[0][i])

    
def writeOut(A,B,Cin,Cout,Sum,Total):
    
    outF = open("RCA_4_Bit.txt", "w")
    
    #giving values to text required for .txt doc
    l ='A';
    m ='B';
    n ='Cin';
    o ='Cout';
    p = 'Sum';
    q ='Total';


    #formatting .txt doc
    outF.write(f'{l:^4} {m:^4} {n:^4} {o:^4} {p:^4} {q:^4} \n')
    temp1 = []
    temp2 = []    
    temp3 = []    
    temp4 = []  
    temp5 = []
    temp6 = []
    for i in range(np.size(A)):
 
        temp1.append(str(A[0][i]))
        temp2.append(str(B[0][i]))
        temp3.append(str(Cin[0][i]))
        temp4.append(str(Cout[i]))
        temp5.append(str(Sum[i]))
        temp6.append(str(Total[i]))

  
    for i in range(np.size(A)):
        outF.write(f'{temp1[i]:^4} {temp2[i]:^4} {temp3[i]:^4} {temp4[i]:^4} {temp5[i]:^4} {temp6[i]:^4} \n')

    
    outF.close()
    
    return


writeOut(A,B,Cin,Cout,b_10_convert,Total)
````
